```python
def cal_sal(country):
    if country == 'USA':
        return 100+100*.3
    elif country == 'Germany':
        return 100+100*.25
    elif country == 'Canada':
        return 100+100*.20
    elif country =='India':
        return 100*100*.15
    elif country=='UK':
        return 100+.23*100
    else:
        return 100+100*.05
```

```python
mydf['salary'] = mydf['country'].apply(cal_sal)
```
