```python
df_csv.loc['2017-07-01', ['High','Low']].resample('D').agg('max','min') #agg({'High':'max','Low':'min'})
```

```python
#this is not right method to get the various aggregated result as it will give you woring becauase row of comnines columns and it will give you result based on the one value mentioned last
#use aggregated value in dictionary to get the right numbers
dl= df_csv.loc['2017-07-01', ['High','Low']].resample('D').agg('max','min')
#df_csv[['High','Low']].resample('D').agg('max','min')
dl
```

```python
dl.rename(columns={'High':'NonD-High', 'Low':'Non-D-Low'}, inplace=True)
```

```python
#combine two dataframe side by side - axis value decide how to join DF's axis(0-row, 1-columns)or(False=rows, True=Columns)
ndh = pd.concat([dh,dl], axis=True, sort=False)
ndh
```

```python
#obser the diff in output
pd.concat([dh,dl], axis=False, sort=False)
```
